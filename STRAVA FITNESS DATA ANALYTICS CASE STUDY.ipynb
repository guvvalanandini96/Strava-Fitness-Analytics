## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# ==========================================
# Data Wrangling
# ==========================================

# Convert date columns into datetime format

daily_activity['ActivityDate'] = pd.to_datetime(daily_activity['ActivityDate'])

daily_calories['ActivityDay'] = pd.to_datetime(daily_calories['ActivityDay'])

daily_intensities['ActivityDay'] = pd.to_datetime(daily_intensities['ActivityDay'])

daily_steps['ActivityDay'] = pd.to_datetime(daily_steps['ActivityDay'])

sleep_day['SleepDay'] = pd.to_datetime(sleep_day['SleepDay'])

weight_log['Date'] = pd.to_datetime(weight_log['Date'])


# Remove duplicate records

daily_activity.drop_duplicates(inplace=True)
daily_calories.drop_duplicates(inplace=True)
daily_intensities.drop_duplicates(inplace=True)
daily_steps.drop_duplicates(inplace=True)
sleep_day.drop_duplicates(inplace=True)
weight_log.drop_duplicates(inplace=True)


# Fill missing values

weight_log['Fat'] = weight_log['Fat'].fillna(weight_log['Fat'].median())


# Merge datasets

fitness_data = daily_activity.merge(
    daily_calories,
    left_on=['Id','ActivityDate'],
    right_on=['Id','ActivityDay'],
    how='left'
)

fitness_data = fitness_data.merge(
    daily_steps,
    left_on=['Id','ActivityDate'],
    right_on=['Id','ActivityDay'],
    how='left'
)

fitness_data = fitness_data.merge(
    daily_intensities,
    left_on=['Id','ActivityDate'],
    right_on=['Id','ActivityDay'],
    how='left'
)

print(fitness_data.shape)

fitness_data.head()
````

### What all manipulations have you done and insights you found?



#### Data Manipulations Performed

The following data preparation and cleaning steps were performed to make the datasets ready for analysis:

- Imported all six fitness datasets into Python using the Pandas library.
- Converted all date-related columns to the appropriate `datetime` format for accurate time-based analysis.
- Examined the structure, data types, and summary statistics of each dataset.
- Checked for duplicate records and removed them to maintain data consistency.
- Identified missing values across all datasets and handled them appropriately. Missing values in the **Fat** column of the Weight Log dataset were replaced using the median value.
- Verified the uniqueness of user IDs and other important variables.
- Merged the Daily Activity, Daily Calories, Daily Steps, and Daily Intensities datasets using the common **Id** and **Activity Date** columns to create a consolidated dataset for analysis.
- Prepared the cleaned dataset for exploratory data analysis (EDA) and visualization.

#### Insights Found

- The six datasets are interconnected through the common **Id** column, allowing them to be combined for comprehensive analysis.
- Most datasets contain complete and consistent records with very few missing values.
- The Weight Log dataset contains fewer observations than the activity datasets because not all users regularly record their weight and body fat.
- Daily Activity provides the most comprehensive information about users' fitness behavior, including steps, calories, distance, and activity duration.
- Converting date columns into the correct format enables trend analysis over time.
- The merged dataset provides a complete view of users' daily fitness activities, making it suitable for identifying behavioral patterns, evaluating activity levels, analyzing sleep habits, and generating business recommendations for Bellabeat.

### Chart 12 - Average Daily Distance by Day of the Week

````python
# ==========================================
# Chart - 12
# Average Daily Distance by Day of the Week
# ==========================================

# Create Day column if not already created
daily_activity['Day'] = pd.to_datetime(
    daily_activity['ActivityDate']
).dt.day_name()

# Arrange days in order
day_order = [
    'Monday',
    'Tuesday',
    'Wednesday',
    'Thursday',
    'Friday',
    'Saturday',
    'Sunday'
]

avg_distance = daily_activity.groupby('Day')['TotalDistance'].mean().reindex(day_order)

plt.figure(figsize=(10,6))

plt.plot(
    avg_distance.index,
    avg_distance.values,
    marker='o',
    linewidth=3,
    color='darkgreen'
)

plt.title("Average Daily Distance by Day of the Week", fontsize=16)
plt.xlabel("Day of Week")
plt.ylabel("Average Distance")

plt.grid(alpha=0.3)

plt.show()
````
